# 3. Create a util function to prepare dictionary based on language name and number of speakers (consider only the columns pertaining to the mother tongue)

```python
def language_dictionary_util_a(df, language_dict):

    # iterate over the rows in dataframe
    for i, row in df.iloc[5:].iterrows():

        # if the entry is not null then the row contains language and number of its speakers
        if(not pd.isnull(row.iloc[3])):
            
            # if the key already exist, add the number of its speakers to the old value, else create a new key-value pair
            if(row.iloc[3] in language_dict):
                # to the dictionary add the language as key and number of its speakers as value
                language_dict[row.iloc[3]] += int(row.iloc[4])
            else:
                language_dict[row.iloc[3]] = int(row.iloc[4])

    return language_dict
```
